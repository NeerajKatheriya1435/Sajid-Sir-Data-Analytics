# Pandas Series Lecture 1

```python
import pandas as pd # alia name

data1=pd.Series([5,3,2,8])

print(data1)
print(type(data1))
```

```python
data2=pd.Series([5,3,2,8],index=["A1","B1","C","D"],dtype=float,name="Python Data")
print(data2)
```

```python
dic={"sub":["phy","chem","math"],"marks":[20,45,89],"rank":[2,5,7]}
ser1=pd.Series(dic)
print(ser1)
```

```python
dic1={11:[34,23,23],12:[20,45,89],31:[2,5,7]}
ser2=pd.Series(dic1)
print(ser2)
```

```python
s1=pd.Series(34,index=[1,2,3,4,5])

print(s1)

print(type(s1))
```

```python
d1=pd.Series(12,index=[1,5,7,8])

d2=pd.Series(15,index=[1,3,6,8])
print(d1)
print(d2)

print(d1+d2)
```

```python
s1=pd.Series(12,index=[1,2,5,7,8,9])
s2=pd.Series(12,index=[1,3,5,7])
print(s1+s2)
```

# Arithmetic operation in DataFrame

```python
d4=pd.DataFrame({"a":[1,2,4,6],"s":[4,6,8,9],"d":[2,3,5,6]})

# d4["sum"]=d4["a"]+d4["d"]+d4["s"]
# d4
```

```python
d4["Multi"]=d4["s"]*d4["d"]

d4["D4>20"]=d4["Multi"]>20
d4
```

```python
d4["Sum"]=d4["a"]+d4["s"]+d4["d"]
d4
d4["Diff(a-b)"]=d4["a"]-d4["s"]
d4
d4["Mul(a*b)"]=d4["a"]*d4["s"]
d4
d4["Div(a/b)"]=d4["a"]/d4["s"]
d4
```
