### Area Graph and Stacked Graph

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# plt.fill_between(x, y)

plt.stackplot(x, y)
plt.show()
```

```python
y1 = [10, 20, 8, 40, 50]
y2 = [5, 15, 25, 35, 45]

plt.fill_between(x, y1, label='Product A')
plt.fill_between(x, y2,alpha=.7, label='Product B')

plt.legend()
plt.show()
```

### Stacked Graph

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]
y3 = [2, 10, 15, 20, 30]

c11=["r","g","b"]
# custom_colors = ['#FF9999', 'skyblue', 'lightgreen']
a11=["First","Second","Third"]

# plt.stackplot(x, y1, y2, y3,labels=a1,colors=c1,baseline="sym")
# plt.stackplot(x, y1, y2, y3,labels=a1,colors=c1)
plt.stackplot(x, y1, y2, y3,colors=c11,labels=a11,baseline="sym")
# plt.stackplot(x, y1, y2, y3,labels=a1)
# plt.stackplot(x, y1, y2, y3)

plt.legend(loc=2)
plt.show()
```

### Step Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 10, 20, 20, 30]

plt.step(x, y,color="r",marker="o",ms=10,mfc="m",label="Growth")
# plt.step(x, y,color="r",marker="o",ms=12,mfc="g")

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Basic Step Plot")
plt.legend(loc=4)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y1 = [5, 5, 10, 10, 15]
y2 = [3, 6, 6, 9, 12]

plt.step(x, y1, where='post', label='Product A',marker="*",ms=10,mfc="m")
plt.step(x, y2, where='post', label='Product B',marker="*",ms=10,mfc="m")

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Multi-Step Graph")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 11)
y = np.array([2, 4, 6, 8, 3, 7, 9, 4, 6, 2])

# Line plot
plt.plot(x, y, label="Data Line", linewidth=2)

# Fill area where y > 5
plt.fill_between(x, y, where=(y > 5),
                 alpha=0.4,
                 interpolate=True,
                 label="y > 5")

# Labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Plot and Fill Between Example")

# Legend and grid
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(10, 8))

# Subplot 1: Line Plot
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

# Subplot 2: Bar Graph
plt.subplot(2, 2, 2)
plt.bar(x, y)
plt.title("Bar Graph")
plt.xlabel("X")
plt.ylabel("Y")

# Subplot 3: Scatter Plot
plt.subplot(2, 2, 3)
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Subplot 4: Histogram
plt.subplot(2, 2, 4)
plt.hist(y, bins=5)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Adjust layout
plt.tight_layout()

# Show all plots
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Plot")

axs[0, 1].bar(x, y)
axs[0, 1].set_title("Bar Graph")

axs[1, 0].scatter(x, y)
axs[1, 0].set_title("Scatter Plot")

axs[1, 1].hist(y, bins=5)
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```
