### MultiPie Charts

```python
import matplotlib.pyplot as plt

sizes1 = [40, 30, 20, 10] 
sizes2 = [34, 67, 26, 78] 
sizes3 = [1] 
labels = ['Rent', 'Food', 'Travel', 'Savings']

color=["r","b","y","g"]

plt.pie(sizes1,labels=labels,autopct='%1.1f%%',radius=1.2)
plt.pie(sizes2,radius=0.58,colors="w")
plt.pie(sizes2,autopct='%1.1f%%',radius=0.5)

plt.title("Monthly Expense Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes1 = [40, 30, 20, 10] 
sizes2 = [1] 
labels = ['Rent', 'Food', 'Travel', 'Savings']
color=["r","b","y","g"]

plt.pie(sizes1,labels=labels,autopct='%1.1f%%',radius=1)
plt.pie(sizes2,radius=0.4,colors="w")

plt.title("Monthly Expense Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes1 = [40, 30, 20, 10] 
labels = ['Rent', 'Food', 'Travel', 'Savings']
color=["r","b","y","g"]

plt.pie(sizes1,labels=labels,autopct='%1.1f%%',radius=1)
cr=plt.Circle(xy=(0,0),radius=0.8,facecolor="w")
plt.gca().add_artist(cr)

plt.title("Monthly Expense Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    wedgeprops={'width': 0.6}
)
plt.show()
```

### Stem Chart

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# plt.stem(x, y,linefmt='--',markerfmt='o',basefmt='k-',orientation="horizontal")
plt.stem(x, y,linefmt=":",markerfmt="r+",basefmt='k-',orientation="horizontal")

plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Stem Plot Example")
plt.show()
```

### MultiBox Plotting

```python
import matplotlib.pyplot as plt

data1 = [10, 20, 25, 30, 35, 40]
data2=[30,56,45,10,89,34]
data3=[20,96,65,10,89,34]

y=[data1,data2,data3]

plt.boxplot(y,labels=["Python","Java","Node.js"],patch_artist=True,showmeans=True,capprops=dict(color="b"),whiskerprops=dict(color="g"))

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.xlabel("<-- Languages -->")
plt.show()
```
