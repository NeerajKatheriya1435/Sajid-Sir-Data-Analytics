```python
import matplotlib.pyplot as plt
import numpy as np


langauges=["JavaScript","Python","Java","C/C++","Dart"]

growthLevel1=[70,80,85,67,50] # 2024
growthLevel2=[54,87,45,62,88] # 2025

colors=["r","y","b","g","c"]
width=0.4

p1=np.arange(len(langauges))
p2=[j+width for j in p1]

# plt.bar(langauges,growthLevel,color=colors,edgecolor="b",linewidth=4,linestyle=":",alpha=0.4,width=0.4)
plt.bar(p1,growthLevel1,width,edgecolor="b",label="Language1") # 1,2,3,4,5 0.4
plt.bar(p2,growthLevel2,width,edgecolor="g",label="Language2") # 1.4,2.4,3.4,4.4,5.4

plt.legend()
plt.xlabel("<-- Languages -->")
plt.ylabel("<-- Growth -->")
plt.title("Language Popularity Bar Garph")
# plt.xticks(p1+width/2,langauges,rotation=20)
plt.xticks(p1+width/2,langauges)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
temperature = [30, 40,35,23,18,32]
day = [1, 2, 3, 4, 5, 6]
# colors=["r","g","b","g","c","y"]
colors=[23,56,34,12,23,42]
sizes=[300,400,350,230,180,320]
fig, ax = plt.subplots()

# Scatter plot
# plt.scatter(day, temperature)

plt.scatter(day, temperature,c=colors,s=sizes,alpha=0.6,marker="*",edgecolors="r",linewidths=2)
# plt.scatter(day, temperature,c=colors,s=sizes,cmap="berlin",marker="*",alpha=0.5)
# plt.scatter(day, temperature,c=colors,s=sizes)
# plt.scatter(day, temperature)

plt.xlabel("Days")
plt.ylabel("Temperature in Degree")
plt.title("Scatter Plot: Day vs Temperature")
plt.colorbar()

ax.set_facecolor('cyan')
plt.show()
```
